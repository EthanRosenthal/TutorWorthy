```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import seaborn as sns
import pickle
import itertools
import json
import re
import sklearn

from sklearn import cross_validation
from sklearn import tree
from sklearn import svm
from sklearn import ensemble
from sklearn import neighbors
from sklearn import linear_model
from sklearn import metrics
from sklearn import preprocessing
from sklearn import grid_search
from sklearn.externals import joblib

sns.set()
```

```python
cd ../data/
```

```python
possible_subjects = pickle.load(open('possible_subjects.pkl', 'rb'))
```

```python
feature_labels = [ 'EdD', 'Enrolled', \
                            'Graduate_Coursework', 'J.D.', 'MBA', 'MD', \
                            'MEd', 'Masters', 'Other', 'PhD', \
                            'avg_review_length', 'badge_hours', \
                            'days_since_last_review', 'has_badge', \
                            'has_rating', 'lat', 'lon', \
                            'number_of_ratings', 'number_of_reviews', \
                            'profile_picture', 'rating', 'zip_code', \
                            'zip_radius', 'rating_x_number_of_ratings']

feature_labels.extend(possible_subjects)
```

```python
cd sklearn/
```

```python
rf = joblib.load('rf.joblib')
```

```python
scaler_X = pickle.load(open('scaler_X.pkl'))
scaler_y = pickle.load(open('scaler_y.pkl'))
X = pickle.load(open('X.pkl'))
y = pickle.load(open('y.pkl'))
```

```python
# Save for later
X_train, X_final_test, y_train, y_final_test = \
    cross_validation.train_test_split(X, \
                                      y, \
                                      test_size=.2, \
                                      random_state=19)

y_train_trans = scaler_y.inverse_transform(y_train)
y_pred_trans = scaler_y.inverse_transform(rf.predict(X_train))
```

```python
plt.figure(figsize=(16, 12))

plt.scatter(range(y_pred_trans.shape[0]), y_pred_trans, label='predictions', c='#348ABD', alpha=0.4)
plt.scatter(range(y_train_trans.shape[0]), y_train_trans, label='actual values', c='#A60628', alpha=0.4)
plt.ylim([y_train_trans.min(), y_pred_trans.max()])
plt.xlim([0, y_pred_trans.shape[0]])
plt.legend()
```

```python
clf_best_estimator = rf

# Get Feature Importance from the classifier
feature_importance = clf_best_estimator.feature_importances_
# Normalize The Features
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
plt.figure(figsize=(14, 26))
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, np.asanyarray(feature_labels)[sorted_idx])
plt.xlabel('Relative Importance', fontsize=20)
plt.title('Feature Importance', fontsize=20)
plt.show()
```
