```python
c.df = c.df[c.df['has_badge']==1]
zip_codes = c.df['zip_code'].unique()
zip_mean = []
zip_std = []
lat = []
lon = []
for zc in zip_codes:
    zip_mean.append(c.df[c.df['zip_code']==zc]['hourly_rate'].mean())
    zip_std.append(c.df[c.df['zip_code']==zc]['hourly_rate'].std())
    lat.append(c.df[c.df['zip_code']==zc]['lat'].values[0])
    lon.append(c.df[c.df['zip_code']==zc]['lon'].values[0])
    
```
