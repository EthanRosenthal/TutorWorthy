```python
def kfold_cv(X, y, kf, clf_reg, **kwargs):
    y_pred = y.copy()
    for ii, jj in kf:
        X_train, X_test = X[ii], X[jj]
        y_train = y[ii]
        clf = clf_reg(**kwargs)
        clf.fit(X_train,y_train)
        y_pred[jj] = clf.predict(X_test)
    return y_pred
```
