## Hmmm, it seems that there are some zip codes missing from pyzipcode, like 10065

```python
from pyzipcode import ZipCodeDatabase
zcdb = ZipCodeDatabase()
```

```python

def get_lat_lon(row, zcdb):
    
    try:
        zip_code = zcdb[int(row)]
        return Series({'lat':zip_code.latitude, 'lon':zip_code.longitude})
    except:
        return Series({'lat':0,'lon':0})
    
lat_lon = d.df['zip_code'].apply(lambda x: get_lat_lon(x, zcdb))
lat_lon = pd.concat([lat_lon, d.df], axis=1)

def convert_radius(row):
    try:
        radius = int(row)
    except:
        radius = 0
    return radius

lat_lon['zip_radius'] = lat_lon['zip_radius'].apply(lambda x: convert_radius(x))
```

```python
lat_lon[(lat_lon['lat']!=0) & (lat_lon['hourly_rate']<150) & (lat_lon['zip_radius']>0)].plot(kind='scatter',x='lon',y='lat', \
                                                                                             c='hourly_rate', s=lat_lon['zip_radius']*20, \
                                                                                             alpha=.1, cmap='jet')
plt.axis([-74.05, -73.85, 40.68, 40.9])
```
