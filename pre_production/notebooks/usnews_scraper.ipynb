```python
import urllib2
from BeautifulSoup import BeautifulSoup
import re
from HTMLParser import HTMLParser
```

```python
base_url = 'http://colleges.usnews.rankingsandreviews.com/best-colleges/rankings/national-universities/data/page+'
page = 1
```

```python
current_url = base_url + str(page)
```

```python
ufile = urllib2.urlopen(current_url)
soup = BeautifulSoup(ufile)
```

```python
rows = soup.findAll('tr', {'valign':'top'})
```

```python
numbers_regex = re.compile('\d+')
words_regex = re.compile('\w+')
```

```python
rank = []
school = []
for row in rows:
    rank.append(re.findall(numbers_regex, \
                           row.find('span').renderContents())[0])
    current_school = row.find('a', {'class':'school-name'}).contents[0]
    current_school = BeautifulSoup(current_school, convertEntities=BeautifulSoup.HTML_ENTITIES)
    current_school = ''.join([i if ord(i)<128 else ' ' for i in current_school.contents[0]])
    current_school = ' '.join(word.lower() for word in current_school.split(' '))
    school.append(current_school)
```

```python
usnews = dict(zip(school, rank))
```

```python
url = 'http://en.wikipedia.org/wiki/List_of_colloquial_names_for_universities_and_colleges_in_the_United_States'
```

```python
soup = BeautifulSoup(urllib2.urlopen(url))
```

```python
soup.findAll('li')
```

Start at the 28th index. Go until YU (number 405).

```python
names = soup.findAll('li')
```

```python
names[50].renderContents()
```

```python
names[50].findAll('a')[0].renderContents()
```

```python
import itertools
school_aliases = {}
for name in names[28:405]:
    alias = name.renderContents().split('-')[0]
    schools = name.findAll('a')
    for school in schools:
        school = school.renderContents()
        if 'or' in alias:
            alias = alias.split(' or ')
        if school_aliases.has_key(school):
            school_aliases[school].append(alias)
        else:
            school_aliases[school] = [alias]

# Get rid of nested lists from 'or' split
for k, v in school_aliases.iteritems():
    if any(isinstance(x, list) for x in v):
        school_aliases[k] = v[0]


    
```
